```python
import pandas as pd


# Function to calculate and print the total number of trains
def total_number_of_trains(df):
    total_trains = df.shape[0]
    print("Total number of trains:", total_trains)
    return total_trains

# Function to calculate and print the count of unique source and destination stations
def unique_stations(df):
    unique_source_stations = df['Source_Station_Name'].nunique()
    unique_destination_stations = df['Destination_Station_Name'].nunique()
    print("Unique source stations:", unique_source_stations)
    print("Unique destination stations:", unique_destination_stations)
    return unique_source_stations, unique_destination_stations

# Function to find and print the most common source and destination stations
def most_common_stations(df):
    most_common_source = df['Source_Station_Name'].mode().iloc[0]
    most_common_destination = df['Destination_Station_Name'].mode().iloc[0]
    print("Most common source station:", most_common_source)
    print("Most common destination station:", most_common_destination)
    return most_common_source, most_common_destination

# Call the functions and print the results
total_trains = total_number_of_trains(df)
unique_source_stations, unique_destination_stations = unique_stations(df)
most_common_source, most_common_destination = most_common_stations(df)
```

```python
# Function to filter the dataset to show trains that operate on specific days
def filter_by_day(df, day):
    # Assuming 'days' column contains information about operating days and needs to be checked for specific day
    filtered_df = df[df['days'].str.contains(day, case=False, na=False)]
    print(f"Trains operating on {day}:")
    print(filtered_df)
    return filtered_df

# Function to extract and create a new dataframe for trains that start from a specific station
def trains_from_station(df, station):
    filtered_df = df[df['Source_Station_Name'].str.upper() == station.upper()]
    print(f"Trains starting from {station}:")
    print(filtered_df)
    return filtered_df

# Example usage
saturday_trains = filter_by_day(df, 'Saturday')
mumbai_trains = trains_from_station(df, 'CST-MUMBAI')
```

Task 2.2: Grouping and Aggregation

Write a function to group the data by source station and print the count of trains
originating from each station.

Write a function to calculate and print the average number of trains per day for each
source station.

```python
import pandas as pd

# Function to group the data by source station and print the count of trains originating from each station
def count_trains_by_source_station(df):
    grouped_df = df.groupby('Source_Station_Name').size().reset_index(name='Train_Count')
    print("Count of trains originating from each source station:")
    print(grouped_df)
    return grouped_df

# Function to calculate and print the average number of trains per day for each source station
def average_trains_per_day(df):
    # Assuming the 'days' column contains a string listing operating days, e.g., "Monday, Tuesday, Wednesday"
    # Convert it to a list of days
    df['Days_List'] = df['days'].str.split(', ')
    # Calculate the number of operating days
    df['Number_of_Days'] = df['Days_List'].apply(len)
    # Calculate the average number of trains per day for each source station
    grouped_avg_df = df.groupby('Source_Station_Name')['Number_of_Days'].mean().reset_index(name='Average_Trains_Per_Day')
    print("Average number of trains per day for each source station:")
    print(grouped_avg_df)
    return grouped_avg_df

# Example usage
train_counts_by_station = count_trains_by_source_station(df)
average_trains = average_trains_per_day(df)
```

Task 2.3: Data Transformation

Write a function to add a new column categorizing trains based on their operating days
(e.g., 'Weekday', 'Weekend').

```python
def categorize_trains_by_days(df):
    # Define the lists of weekday and weekend days
    weekdays = {'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'}
    weekends = {'Saturday', 'Sunday'}

    # Function to categorize the days
    def categorize_days(days_list):
        days_set = set(days_list)
        if days_set & weekends:
            if days_set & weekdays:
                return 'Both'
            else:
                return 'Weekend'
        else:
            return 'Weekday'

    # Apply the function to the 'Days_List' column
    df['Days_Category'] = df['Days_List'].apply(categorize_days)

    return df

# Example usage
df['Days_List'] = df['days'].str.split(', ')
df = categorize_trains_by_days(df)

# Display the first 10 rows to check the new column
print("\nFirst 10 rows with Days_Category:")
print(df.head(10))
```

Task 3.1: Pattern Identification

Write a function to identify and return the distribution of train journeys throughout the
week.

Write a function to identify patterns or trends in train operations based on source and
destination stations.


```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

# Function to identify and return the distribution of train journeys throughout the week
def distribution_of_train_journeys(df):
    # Flatten the list of days into a single list
    all_days = [day for days in df['Days_List'] for day in days]
    day_counts = Counter(all_days)
    distribution_df = pd.DataFrame(day_counts.items(), columns=['Day', 'Count']).sort_values(by='Count', ascending=False)

    print("Distribution of train journeys throughout the week:")
    print(distribution_df)

    # Plotting the distribution
    plt.figure(figsize=(10, 6))
    plt.bar(distribution_df['Day'], distribution_df['Count'], color='skyblue')
    plt.xlabel('Day of the Week')
    plt.ylabel('Number of Train Journeys')
    plt.title('Distribution of Train Journeys Throughout the Week')
    plt.show()

    return distribution_df

# Function to identify patterns or trends in train operations based on source and destination stations
def patterns_in_train_operations(df):
    source_counts = df['Source_Station_Name'].value_counts()
    destination_counts = df['Destination_Station_Name'].value_counts()

    print("Patterns or trends in train operations based on source stations:")
    print(source_counts)

    print("\nPatterns or trends in train operations based on destination stations:")
    print(destination_counts)

    # Plotting source station counts
    plt.figure(figsize=(12, 6))
    source_counts.head(10).plot(kind='bar', color='coral')
    plt.xlabel('Source Station')
    plt.ylabel('Number of Trains')
    plt.title('Top 10 Source Stations by Number of Trains')
    plt.show()

    # Plotting destination station counts
    plt.figure(figsize=(12, 6))
    destination_counts.head(10).plot(kind='bar', color='lightgreen')
    plt.xlabel('Destination Station')
    plt.ylabel('Number of Trains')
    plt.title('Top 10 Destination Stations by Number of Trains')
    plt.show()

    return source_counts, destination_counts

# Example
distribution_df = distribution_of_train_journeys(df)
source_counts, destination_counts = patterns_in_train_operations(df)
```

Task 3.2: Correlation and Insights

Write a function to explore and return any correlation between the number of trains and
specific days of the week.



Write a function to explore and return any correlation between the number of trains and
specific days of the week.

```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

# Function to calculate the number of trains operating on each day
def calculate_trains_per_day(df):
    # Flatten the list of days into a single list
    all_days = [day for days in df['Days_List'] for day in days]
    day_counts = Counter(all_days)
    day_counts_df = pd.DataFrame(day_counts.items(), columns=['Day', 'Train_Count']).sort_values(by='Train_Count', ascending=False)

    print("Number of trains operating on each day:")
    print(day_counts_df)

    return day_counts_df

# Example usage
trains_per_day_df = calculate_trains_per_day(df)
```

```python
import seaborn as sns

# Function to analyze correlation
def analyze_correlation(trains_per_day_df):
    # Convert day names to numeric values for correlation analysis
    days_numeric = {
        'Monday': 1, 'Tuesday': 2, 'Wednesday': 3, 'Thursday': 4, 'Friday': 5,
        'Saturday': 6, 'Sunday': 7
    }
    trains_per_day_df['Day_Numeric'] = trains_per_day_df['Day'].map(days_numeric)

    # Calculate correlation
    correlation = trains_per_day_df['Train_Count'].corr(trains_per_day_df['Day_Numeric'])
    print(f"Correlation between the number of trains and days of the week: {correlation:.2f}")

    # Plotting the correlation
    plt.figure(figsize=(10, 6))
    sns.regplot(x='Day_Numeric', y='Train_Count', data=trains_per_day_df)
    plt.xlabel('Day of the Week (Numeric)')
    plt.ylabel('Number of Trains')
    plt.title('Correlation between Number of Trains and Days of the Week')
    plt.show()

    return correlation

# Example usage
correlation = analyze_correlation(trains_per_day_df)
```

Write a script to print insights and recommendations based on the correlation analysis.


```python
# Function to print insights and recommendations
def print_insights(correlation):
    insights = ""
    recommendations = ""

    if correlation > 0.5:
        insights = "There is a strong positive correlation between the number of trains and specific days of the week."
        recommendations = (
            "Consider increasing train services on days with higher demand to optimize the schedule. "
            "Ensure that resources are adequately allocated to handle the higher traffic on these days."
        )
    elif correlation < -0.5:
        insights = "There is a strong negative correlation between the number of trains and specific days of the week."
        recommendations = (
            "Investigate the reasons for the lower number of trains on certain days. "
            "Consider redistributing services from less busy days to busier ones to balance the load."
        )
    else:
        insights = "There is no significant correlation between the number of trains and specific days of the week."
        recommendations = (
            "The train schedule appears to be well-balanced throughout the week. "
            "Continue monitoring the schedule to ensure it meets passenger demand efficiently."
        )

    print("Insights:")
    print(insights)
    print("\nRecommendations:")
    print(recommendations)

    return insights, recommendations

# Example usage
insights, recommendations = print_insights(correlation)
```
